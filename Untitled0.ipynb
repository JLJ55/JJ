```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet

# Load the dataset
df = pd.read_csv("/content/sales.csv")

# Data Cleaning
print("Initial dataset info:")
print(df.info())
```

```python
# Handle missing values
df.fillna(0, inplace=True)  # Replace NaN values with 0
df['date'] = pd.to_datetime(df['date'])  # Convert 'date' column to datetime

# Drop duplicates
df.drop_duplicates(inplace=True)

# Ensure necessary columns exist
if 'revenue' not in df.columns and 'sales' in df.columns and 'price' in df.columns:
    df['revenue'] = df['sales'] * df['price']

print("\nCleaned dataset info:")
print(df.info())
```

```python

# Sales and revenue by date
daily_data = df.groupby('date')[['sales', 'revenue']].sum().reset_index()

# Data by year
df['year'] = df['date'].dt.year
yearly_data = df.groupby('year')[['sales', 'revenue']].sum().reset_index()

# Plot sales over time
plt.figure(figsize=(12, 6))
plt.plot(daily_data['date'], daily_data['sales'], label='Daily Sales', color='blue')
plt.title('Sales Over Time')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()

# Plot total revenue by year (Bar Chart)
plt.figure(figsize=(12, 6))
plt.bar(yearly_data['year'], yearly_data['revenue'], color='skyblue')
plt.title('Total Revenue by Year')
plt.xlabel('Year')
plt.ylabel('Revenue')
plt.grid(axis='y')
plt.show()
```

```python

# Aggregate sales by month for forecasting
monthly_data = df.groupby(pd.Grouper(key='date', freq='M'))['sales'].sum().reset_index()

# SARIMA Model
sarima_model = SARIMAX(monthly_data['sales'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_result = sarima_model.fit(disp=False)

# Forecast for the next 12 months
sarima_forecast = sarima_result.get_forecast(steps=12)
sarima_predicted = sarima_forecast.predicted_mean

# Prophet Model
prophet_df = monthly_data.rename(columns={'date': 'ds', 'sales': 'y'})
prophet_model = Prophet()
prophet_model.fit(prophet_df)

# Future dataframe and prediction
future = prophet_model.make_future_dataframe(periods=12, freq='M')
forecast = prophet_model.predict(future)

# Plot SARIMA Forecast
plt.figure(figsize=(12, 6))
plt.plot(monthly_data['date'], monthly_data['sales'], label='Historical Sales')
plt.plot(sarima_predicted.index, sarima_predicted.values, label='SARIMA Forecast', color='red')
plt.title('SARIMA Forecast')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid()
plt.show()

# Plot Prophet Forecast
prophet_model.plot(forecast)
plt.title('Prophet Forecast')
plt.show()
```

```python
# Top 5 products by total revenue
top_products = df.groupby('product_id')['revenue'].sum().sort_values(ascending=False).head(5)

# Plot top 5 products by revenue
plt.figure(figsize=(12, 6))
sns.barplot(x=top_products.index, y=top_products.values, palette='viridis')
plt.title('Top 5 Products by Revenue')
plt.xlabel('Product ID')
plt.ylabel('Revenue')
plt.show()

# Revenue distribution across stores
store_revenue = df.groupby('store_id')['revenue'].sum()

# Plot revenue distribution across stores
plt.figure(figsize=(12, 6))
sns.histplot(store_revenue, bins=20, kde=True, color='purple')
plt.title('Revenue Distribution Across Stores')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.grid()
plt.show()
```

```python
import pandas as pd
import numpy as np
from statsmodels.tsa.statespace.sarimax import SARIMAX
from prophet import Prophet
from sklearn.metrics import mean_absolute_percentage_error

# Ensure the 'date' column is in datetime format
df['date'] = pd.to_datetime(df['date'])

# Group data by month
monthly_data = df.groupby(pd.Grouper(key='date', freq='M'))['sales'].sum().reset_index()

# Split into training and testing sets
train = monthly_data[:-12]  # All but the last 12 months
test = monthly_data[-12:]   # Last 12 months

# Debug: Check for zero sales periods in test set
print("Test data before handling zero sales:\n", test)

# Handle zero values in the test set
non_zero_mask = test['sales'] > 0  # Create a mask for non-zero sales

# SARIMA Model
sarima_model = SARIMAX(train['sales'], order=(1, 1, 1), seasonal_order=(1, 1, 1, 12))
sarima_result = sarima_model.fit(disp=False)

# Forecast using SARIMA
sarima_forecast = sarima_result.get_forecast(steps=12)
sarima_predicted = sarima_forecast.predicted_mean

# Prophet Model
prophet_train = train.rename(columns={'date': 'ds', 'sales': 'y'})
prophet_model = Prophet()
prophet_model.fit(prophet_train)

# Forecast using Prophet
future = prophet_model.make_future_dataframe(periods=12, freq='M')
forecast = prophet_model.predict(future)
prophet_predicted = forecast[-12:]['yhat'].values

# Debug: Print predictions to verify consistency
print("\nSARIMA Predictions:\n", sarima_predicted)
print("\nProphet Predictions:\n", prophet_predicted)

# Calculate MAPE for SARIMA and Prophet, excluding zero sales periods
sarima_mape = mean_absolute_percentage_error(
    test['sales'][non_zero_mask], sarima_predicted[non_zero_mask]
) * 100

prophet_mape = mean_absolute_percentage_error(
    test['sales'][non_zero_mask], prophet_predicted[non_zero_mask]
) * 100

# Display Results
print(f"\nSARIMA Model MAPE: {sarima_mape:.2f}%")
print(f"Prophet Model MAPE: {prophet_mape:.2f}%")

# Check if both models passed the success criterion (MAPE ≤ 10%)
success_criteria = 10  # Success threshold for MAPE
sarima_success = sarima_mape <= success_criteria
prophet_success = prophet_mape <= success_criteria

if sarima_success and prophet_success:
    print("Both models achieved high accuracy, with MAPE values surpassing the success criteria of ≤10%.")
elif sarima_success:
    print("SARIMA model achieved high accuracy, surpassing the success criteria of ≤10%.")
elif prophet_success:
    print("Prophet model achieved high accuracy, surpassing the success criteria of ≤10%.")
else:
    print("Neither model met the success criteria of ≤10%.")
```
